#### Handling Imbalanced datasets using above data


```python
def ANN(X_trian,y_trian,X_test,y_test,loss,weights):
    model = tf.keras.Sequential([
    tf.keras.layers.Dense(20, activation='relu', input_shape=(26,)),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
    ])


    model.compile(optimizer='adam', 
                loss=loss, 
                metrics=['accuracy'])
    
    if weights == -1:
        model.fit(X_trian,y_train,epochs=100)
    else:
        model.fit(X_trian,y_test,epochs=100,weights=weights)

    print(model.evaluate(X_test,y_test))

    y_preds = model.predict(X_test)
    y_preds = np.round(y_preds)


    print('Classification Report: \n',classification_report(y_test,y_preds))

    return y_preds
```

```python
y_preds = ANN(X_train,y_train,X_test,y_test,'binary_crossentropy',-1)
```

### Downscaling the datasets to minority class

```python
count_class_0, count_class_1 = df1.Churn.value_counts()

df_class_0 = df2[df2['Churn']==0]
df_class_1 = df2[df2['Churn']==1]
```

```python
df_class_0.shape
```

```python
df_class_1.shape
```

```python
df_class_0_under = df_class_0.sample(count_class_1)
df_test_under=pd.concat([df_class_0_under, df_class_1],axis = 0)
print("Random under-sampling")
print(df_test_under['Churn'].value_counts())
```

```python
df_test_under.head()
```

```python
X = df_test_under.drop(['Churn'],axis=1)
y = df_test_under['Churn']


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=18,stratify=y)
```

```python
y_train.value_counts()
```

```python
y_preds = ANN(X_train,y_train,X_test,y_test,'binary_crossentropy',-1)
```

### Method:3 SMOTE

```python
X = df2.drop('Churn',axis=1)
y= df2['Churn']
```

```python
y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy='minority')

X_sm,y_sm = smote.fit_resample(X,y)

y_sm.value_counts()
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.2, random_state=18,stratify=y_sm)
```

```python
y_train.value_counts()
```

```python
y_preds = ANN(X_train,y_train,X_test,y_test,'binary_crossentropy',-1)
```

### MOdel 4 Use of Ensamble with undersampling

```python
df2.Churn.value_counts()
```

```python
X = df2.drop('Churn',axis=1)
y= df2['Churn']
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=18,stratify=y)
```

```python
y_train.value_counts(
    
)
```

```python
df3 = X_train.copy()
df3['Churn'] = y_train
```

```python
df3_class0 = df3[df3.Churn==0]
df3_class1 = df3[df3.Churn==1]
```

```python
df3_class0.shape, df3_class1.shape
```

```python
def get_train_batch(df_majority,df_minority,start,end):
    df_train = pd.concat([df_majority[start:end], df_minority], axis=0)

    X_train = df_train.drop('Churn', axis=1)
    y_train = df_train['Churn']

    return X_train, y_train
```

```python
X_train,y_train = get_train_batch(df3_class0, df3_class1,0,1495)
y_pred1 = ANN(X_train,y_train,X_test,y_test,'binary_crossentropy',-1)
```

```python

X_train,y_train = get_train_batch(df3_class0, df3_class1,1495,2990)
y_pred2 = ANN(X_train,y_train,X_test,y_test,'binary_crossentropy',-1)
```

```python
X_train,y_train = get_train_batch(df3_class0, df3_class1,2990,4130)
y_pred3 = ANN(X_train,y_train,X_test,y_test,'binary_crossentropy',-1)
```

```python
y_pred_final = y_pred1.copy()


for i in range(len(y_pred1)):

    n_ones = y_pred1[i] + y_pred2[i] + y_pred3[i]
    if n_ones >1:
        y_pred_final[i] = 1
    else:
        y_pred_final[i] = 0
```

```python
print(classification_report(y_test, y_pred_final
                            ))
```
